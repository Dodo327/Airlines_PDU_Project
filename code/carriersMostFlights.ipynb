```python
import pandas as pd 
import numpy as np
from util.getData import *
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = getData(["UniqueCarrier", "TailNum"])
carriers = getCarriers()
```

```python
df = df.dropna(axis=0)
```

```python
df["count"] = df.groupby("UniqueCarrier")["UniqueCarrier"].transform('count')
df = df[["UniqueCarrier", "count"]].drop_duplicates()
```

```python
df = df.sort_values("count", ascending=False).head(10)
df["count"] /= 1000000
print(df)
```

```python
df.plot(kind="bar", legend=None)
plt.title("Flights per Carrier")
plt.xlabel("Carriers")
plt.ylabel("Number of flights *10^6")
plt.xticks(np.arange(10), list(df.UniqueCarrier), rotation=45)
plt.show()
```
